```python
import ee
import geemap
```

```python
geemap.ee_initialize()
```

### 1. **Display cloudsen12 locations**

```python
regions = ee.FeatureCollection("projects/sat-io/open-datasets/cloudsen12/footprint")

palette = ee.Dictionary({
  'high': 'red',
  'scribble': 'blue',
  'nolabel': 'green',
})

# from FeatureCollection to Image
def addf(feature):
  return feature.set('myStyle', {
    'color': palette.get(feature.get('lbl_typ')),
    'pointSize': 1
  })
image = regions.map(addf).style(styleProperty = 'myStyle')
```

```python
Map = geemap.Map(center=[0, 0], zoom=3)
Map.addLayer(image, {})
Map
```

### 2. **Display a cloudSEN12 ROI**

```python
ic_high = ee.ImageCollection("projects/sat-io/open-datasets/cloudsen12/high")
ic_high_filtered = ic_high.filter(ee.Filter.eq('roi_id', "ROI_0001"))
```

```python
Map = geemap.Map(zoom=12)
Map.centerObject(ic_high_filtered.first())

for index in range(5):
  # Select image  
  img = ee.Image(ic_high_filtered.toList(1, index).get(0))
  roi = img.get("roi_id").getInfo()
  s2id = img.get("s2_id_gee").getInfo()
  
  # s2 image
  S2img = ee.Image("COPERNICUS/S2/%s" % s2id)

  # Send to geemap
  layername = "%s__%s" % (roi, s2id)
  Map.addLayer(S2img, {"bands": ["B4", "B3", "B2"], "min":0 ,"max": 3000}, layername + "_s2")
  Map.addLayer(img, {"min":0, "max":3, "palette": ['00FFFF', 'FFFF00', '00FF00', 'FF0000']}, layername + "_hq")

Map
```
